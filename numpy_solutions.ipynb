```python
#q.Получить сумму всех значений в стартовой матрице mat
mat = np.arange(1,26).reshape(5,5)
print("Сумма значений =", mat.sum())
```

```python
#r.Получить стандартное отклонение в стартовой матрице mat
import numpy as np
mat = np.arange(1,26).reshape(5,5)
print("Стандартное отклонение =", np.std(mat))
```

```python
#s.Получить сумму значений в колонках в стартовой матрице mat
import numpy as np
mat = np.arange(1,26).reshape(5,5)
a = mat[:,0]
print ("Колонка 1", a)
print ("Сумма значений в 1 колонке =", a.sum())
print ("____________________________")

b = mat[:,1]
print ("Колонка 2", b)
print ("Сумма значений в 2 колонке =", b.sum())
print ("____________________________")

c = mat[:,2]
print ("Колонка 3", c)
print ("Сумма значений в 3 колонке =", c.sum())
print ("____________________________")

d = mat[:,3]
print ("Колонка 4", d)
print ("Сумма значений в 4 колонке =", d.sum())
print ("____________________________")

e = mat[:,4]
print ("Колонка 5", e)
print ("Сумма значений в 5 колонке =", e.sum())
print ("____________________________")
```
